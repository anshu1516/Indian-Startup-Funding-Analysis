**Total number of fundings done in each year.**

```python
# extracting year from date
strtup['Year'] = strtup['Date'].dt.year
```

```python
yrwise_fnd_cnt = strtup.groupby('Year')['Amount in USD'].count().reset_index()
yrwise_fnd_cnt
```

**Plot a line graph between year and number of fundings. Take year on x-axis and number of fundings on y-axis.**

```python
sns.lineplot(x='Year',y='Amount in USD',data=yrwise_fnd_cnt,marker='o',c='turquoise')
plt.ylabel('Number of fundings')
plt.title('Number of fundings per Year')
plt.show()
```

**Print year-wise total number of fundings also. Print years in ascending order.**

```python
year = yrwise_fnd_cnt['Year']
ttl_cnt = yrwise_fnd_cnt['Amount in USD']
my_dict = dict(zip(year,ttl_cnt))

for key,value in my_dict.items():
  print(f'The total number of startup funded in {key}: {value}')
```

```python
strtup['Investment Type'].unique()
```

```python
# Correct spelling of investment types are - "Private Equity", "Seed Funding", "Debt Funding", and "Crowd Funding". Keep an eye for any spelling mistake.
strtup['Investment Type'] = np.where(strtup['Investment Type'].isin(['Private Equity Round','Equity','Equity Based Funding','Private Funding',
                                                                     'Private','PrivateEquity','Private\\\\nEquity']),'Private Equity',strtup['Investment Type'])
strtup['Investment Type'] = np.where(strtup['Investment Type'].isin(['pre-series A','Pre-series A','pre-Series A']),'Pre-Series A',strtup['Investment Type'])
strtup['Investment Type'] = np.where(strtup['Investment Type']=='Series B (Extension)','Series B',strtup['Investment Type'])
strtup['Investment Type'] = np.where(strtup['Investment Type'].isin(['Seed','Seed Round','Seed Funding Round','Seed funding','Seed\\\\nFunding']),
                                     'Seed Funding',strtup['Investment Type'])
strtup['Investment Type'] = np.where(strtup['Investment Type'].isin(['Venture Round','Venture - Series Unknown']),'Single Venture',strtup['Investment Type'])
strtup['Investment Type'] = np.where(strtup['Investment Type'].isin(['Angel','Angel Round']),'Angel Funding',strtup['Investment Type'])
strtup['Investment Type'] = np.where(strtup['Investment Type'].isin(['Debt','Debt and Preference capital','Structured Debt','Debt-Funding']),
                                     'Debt Funding',strtup['Investment Type'])
strtup['Investment Type'] = np.where(strtup['Investment Type'].isin(['Seed/ Angel Funding','Seed / Angel Funding','Seed / Angle Funding',
                                                                     'Angel / Seed Funding']),'Seed/Angel Funding',strtup['Investment Type'])
strtup['Investment Type'] = np.where(strtup['Investment Type']=='Crowd funding','Crowd Funding',strtup['Investment Type'])
```

**Print the investment type and percentage of amount funded with 2 decimal places after rounding off.**
